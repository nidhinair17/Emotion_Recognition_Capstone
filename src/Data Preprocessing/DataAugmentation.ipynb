```python
def time_stretch_with_resampling(input_file, output_file, rate=0.8):
    # Load the audio file
    audio, sr = librosa.load(input_file, sr=None)

    # Apply time-stretching
    stretched_audio = librosa.effects.time_stretch(audio, rate=0.8)

    # Resample to original sample rate for better quality
    resampled_audio = librosa.resample(stretched_audio, orig_sr=sr, target_sr=sr)

    # Save the processed audio
    sf.write(output_file, resampled_audio, sr)
```

```python
Data_folder = "C:/Nini/Capstone/AudioFilesSnippets"
Destination_folder = "C:/Nini/Capstone/AudioFilesAugmentedSnippets"
```

```python
for folder in os.listdir(Data_folder):
    folder_path = f'{Data_folder}/{folder}'
    for file in os.listdir(folder_path):
        file_Path = f'{folder_path}/{file}'
        audio, sr = librosa.load(file_Path)
        os.makedirs(f'{Destination_folder}/{folder}',exist_ok=True)
        output_folder = f'{Destination_folder}/{folder}'
        print(output_folder)
        sf.write(f'{output_folder}/{file}',audio,sr)
        file_name = file.split('.')[0]
        time_stretch_with_resampling(file_Path,f'{output_folder}/{file_name}-stretched.wav',rate=0.9)
        
```
