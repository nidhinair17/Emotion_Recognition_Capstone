```python
import pandas as pd
import os

# Load both Excel files
df1 = pd.read_csv("C:/Nini/Capstone/CSV_Files/Speech data/Speech_Train_preprocessed_new.csv")   # Contains .wav and -stretched.wav files
df2 = pd.read_excel("C:/Nini/Capstone/CSV_Files/Facial data/FacialFeatures(New).xlsx")  # Contains .mp4 files

# Get base names from df1 and remove '-stretched' if present
def clean_base_name(path):
    base = os.path.splitext(os.path.basename(str(path)))[0]
    return base.replace('-stretched', '')  # Remove '-stretched' suffix if present

df1['BaseName'] = df1['Path'].apply(clean_base_name)
# Get base names from df2
df2['BaseName'] = df2['video_path'].apply(lambda x: os.path.splitext(os.path.basename(str(x)))[0])

# Create a set of base names from df1
base_name_set = set(df1['BaseName'].dropna().unique())
# print(df2['BaseName'])

# Filter df2 to include only rows whose BaseName is in df1
filtered_df = df2[df2['BaseName'].isin(base_name_set)]
filter_df_test = df2[~df2['BaseName'].isin(base_name_set)]
# Drop the helper column if you don't need it
filtered_df = filtered_df.drop(columns=['BaseName'])

# Save or display the result
# filtered_df.to_excel("filtered_result.xlsx", index=False)
print(filtered_df)
print(filter_df_test)

# filtered_df.drop(['frame'],axis=1, inplace=True)
# filter_df_test.drop(['frame'],axis=1, inplace=True)
filtered_df.to_excel('C:/Nini/Capstone/CSV_Files/Facial data/FacialFeatures_Train_new.xlsx')
filter_df_test.to_excel('C:/Nini/Capstone/CSV_Files/Facial data/FacialFeatures_Test_new.xlsx')

```

```python
import pandas as pd
import os

# Load both files
df1 = pd.read_csv("C:/Nini/Capstone/CSV_Files/Speech data/Speech_Train_preprocessed_new.csv")
df2 = pd.read_excel("C:/Nini/Capstone/CSV_Files/Facial data/FacialFeatures(New).xlsx")

# Filter out rows with '-stretched' in the Path column
df1 = df1[~df1['Path'].str.contains('-stretched', na=False)]

# Get base names from df1
df1['BaseName'] = df1['Path'].apply(lambda x: os.path.splitext(os.path.basename(str(x)))[0])

# Get base names from df2
df2['BaseName'] = df2['video_path'].apply(lambda x: os.path.splitext(os.path.basename(str(x)))[0])

# Create set of base names from df1
base_name_set = set(df1['BaseName'].dropna().unique())

# Filter df2 based on presence in df1 base names
filtered_df = df2[df2['BaseName'].isin(base_name_set)]
filter_df_test = df2[~df2['BaseName'].isin(base_name_set)]

# Drop helper column
filtered_df = filtered_df.drop(columns=['BaseName'])
filter_df_test = filter_df_test.drop(columns=['BaseName'])

# Save outputs
filtered_df.to_excel('C:/Nini/Capstone/CSV_Files/Facial data/FacialFeatures_Train_new.xlsx', index=False)
filter_df_test.to_excel('C:/Nini/Capstone/CSV_Files/Facial data/FacialFeatures_Test_new.xlsx', index=False)
```
